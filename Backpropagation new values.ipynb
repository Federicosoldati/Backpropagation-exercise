```python
#Ignore the warnings
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import tensorflow as tf
import math
```

```python
num_data = np.array([[1,1], [0, -2], [-1, 1]])
target = np.array([[-1], [1], [1]])
```

```python
def get_trainable_variables(graph=tf.compat.v1.compat.v1.get_default_graph()):
    return [v for v in graph.get_collection(tf.compat.v1.GraphKeys.TRAINABLE_VARIABLES)]

n_features = num_data.shape[1]
n_outputs = target.shape[1]
n_hidden = 2

tf.compat.v1.compat.v1.reset_default_graph()

graph = tf.compat.v1.Graph()
with graph.as_default():
    with tf.compat.v1.compat.v1.name_scope('Placeholders'):
        X = tf.compat.v1.placeholder('float', shape=[None, n_features])
        y = tf.compat.v1.placeholder('float', shape=[None, n_outputs])

    with tf.compat.v1.name_scope("First_Layer"):
        W_fc1 = tf.compat.v1.get_variable('First_Layer/Hidden_layer_weights', initializer=tf.compat.v1.constant(np.array([[0.2, -0.7], [-1.2, -0.7]]), dtype=tf.compat.v1.float32))
        b_fc1 = tf.compat.v1.get_variable('First_Layer/Biases', initializer=tf.compat.v1.constant(np.array([0.4, 0.7]), dtype=tf.compat.v1.float32))
        h_fc1 = tf.compat.v1.nn.sigmoid(tf.compat.v1.matmul(X, W_fc1) + b_fc1)

    with tf.compat.v1.name_scope("Output_Layer"):
        W_fc2 = tf.compat.v1.get_variable('Output_Layer/Output_layer_weights', initializer=tf.compat.v1.constant(np.array([[-0.5], [2.3]]), dtype=tf.compat.v1.float32))
        b_fc2 = tf.compat.v1.get_variable('Output_Layer/Biases', initializer=tf.compat.v1.constant(np.array([0.4]), dtype=tf.compat.v1.float32))
        y_pred = tf.compat.v1.cast(tf.compat.v1.matmul(h_fc1, W_fc2) + b_fc2, dtype=tf.compat.v1.float32)

    with tf.compat.v1.name_scope("Loss"):
        loss = tf.compat.v1.losses.mean_squared_error(labels=y, predictions=y_pred)

    with tf.compat.v1.name_scope('Train'):
        optimizer = tf.compat.v1.train.GradientDescentOptimizer(1)
        grads_and_vars = optimizer.compute_gradients(loss)
        trainer = optimizer.apply_gradients(grads_and_vars)

    with tf.compat.v1.name_scope("Init"):
        global_variables_init = tf.compat.v1.global_variables_initializer()

get_trainable_variables(graph=graph)

with tf.compat.v1.Session(graph=graph) as sess:
    global_variables_init.run()
    tf.compat.v1.get_default_graph().finalize()
    print("Initialized")

    print("Variables before training")
    old_var = {}
    for var in tf.compat.v1.global_variables():
        old_var[var.name] = sess.run(var)
    print(old_var)
    print("\n\n")

    # No mini-batches here

    _, loss_val, h_fc1_val, grads_and_vars_val, y_pred_val = sess.run([trainer, loss, h_fc1, grads_and_vars, y_pred], feed_dict={X: num_data, y: target})
    
    print("Loss: {}".format(loss_val))
    print("Prediction: {}".format(y_pred_val))
    print("Hidden layer forward prop: {}".format(h_fc1_val))
    print("\n\n")
    print(grads_and_vars_val)

    print("Optimization Finished!")
    print('\n\n')

    print("Variables after training")
    new_var = {}
    for var in tf.compat.v1.global_variables():
        new_var[var.name] = sess.run(var)
    print(new_var)

```
